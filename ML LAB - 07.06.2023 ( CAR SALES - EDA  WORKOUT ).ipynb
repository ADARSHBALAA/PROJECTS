### MULTI VERIANT PLOTTING

##### LINE CHART

```python
data.sort_values(by='Manufacturer').plot.line('Manufacturer','Sales_in_thousands');
```

```python
data.sort_values(by='Vehicle_type').plot.line('Vehicle_type','Price_in_thousands',figsize=(8,3),cmap='turbo_r');
```

#### BAR PLOT

```python
data.Fuel_efficiency.value_counts().plot.bar();
```

```python
data.Engine_size.value_counts().plot.bar(figsize=(6,3),cmap='inferno');
```

```python
data.Manufacturer.value_counts().plot.bar(figsize=(12,4),cmap='turbo_r');
```

### AREA CHART

```python
data.plot.area('Manufacturer','resale_value');
```

```python
data.plot.area('Model','resale_value',figsize=(12,3),cmap='Pastel2');
```

### HORIZONTAL BAR CHART

```python
data.sort_values(by='Manufacturer')[:10].plot.barh('Manufacturer','Power_perf_factor',figsize=(12,4),cmap='Pastel2')
```

```python
data.sort_values(by='Model')[:20].plot.barh('Model','Wheelbase',figsize=(15,8),cmap='inferno')
```

### SCATTER PLOT

```python
data.sort_values(by='Power_perf_factor')[:100].plot.scatter('Power_perf_factor','Fuel_efficiency',figsize=(4,4))
```

```python
data.sort_values(by='Sales_in_thousands')[:100].plot.scatter('Engine_size','Horsepower',figsize=(4,4));
```
